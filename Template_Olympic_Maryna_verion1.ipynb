#### Section 1: Missing values

First, we identify which columns have missing data and how much.

```python
# TODO: Handle missing values

# Count missing values
missing = df.isnull().sum()
missing_pct = (df.isnull().sum() / len(df) * 100).round(1)

missing_summary = pd.DataFrame({
    "Missing_Count": missing,
    "Missing_Percent": missing_pct
})

print("Missing values summary:")
print(missing_summary[missing_summary['Missing_Count'] > 0].sort_values("Missing_Count", ascending=False))

```

**Analysis of missing values**:

- **notes** (98.1%): It is fine, notes are only provided in rare cases
- **Medal** (85.3%): This is expected - most athletes do not win medals
- **Weight** (23.2%): Weight data is incomplete
- **Height** (22.2%): Height data is incomplete
- **Age** (3.5%): Personal information may not always be recorded
- **region** (0.1%): Some full named of the countries are not provided, which is not a problem

**Decision**: Keep missing values as-is since:
1. We won't be using `Weight` and `Height` columns as we don't have an objective to analyze weight and height.
2. The percentage of missing values for `Age` and `region` columns is relatively low.
3. Analysis functions like `.mean()` handle NaN automatically

For **Medal**, NaN represents "no medal" which is meaningful information.
